```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.info()
```

```python
#delete the cabin cloum cuz we have so many missing values
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.info()
```

```python
df = df.dropna()#Drop the rows where at least element is missing:
df.info()
```

```python
#We will create a new feature (column) called family_size by combining ‘SibSp’(Sibling & Spouse) and ‘Parch’(Parent & Children).
def create_fam_size(df):
    return df["SibSp"] + df["Parch"] + 1

df["family_size"] = create_fam_size(df)
```

```python
df.head(50)
```

```python
def titanic_children(passenger):
    
    age , sex = passenger
    if age <16:
        return 'child'
    else:
        return sex
df['person'] = df[['Age','Sex']].apply(titanic_children,axis=1)
```

```python
df.head(20)
```

```python
df['Alone'] = df.SibSp + df.Parch
# if Alone value is >0 then they are with family else they are Alone
```

```python
df['Alone'].loc[df['Alone']>0] = 'With Family'
df['Alone'].loc[df['Alone'] == 0] = 'Without Family'
```

```python
df.head()
```

```python
df["Survived"].value_counts() #Get a number of who is Survived or not  
```

```python
cleanup_Survived = {"Survived":{"No": 0, "Yes": 1}}
```

```python
df = df.replace(cleanup_Survived) #replace Yes & No with 0 & 1
df.head(10) #print the 1st 5 rows of the data frame
```

```python
df['person'].value_counts() # Get a number of male and female
```

```python
cleanup_Sex = {"person":     {"female": 0, "male": 1 , "child": 3}}
cleanup_Alone = {"Alone":     {"With Family": 1, "Without Family": 0 }}
cleanup_Embarked = {"Embarked":     {"S": 1, "C": 2, "Q" : 3}}
```

```python
df = df.replace(cleanup_Sex)
df = df.replace(cleanup_Alone)
df = df.replace(cleanup_Embarked)
df.head(50)
```
